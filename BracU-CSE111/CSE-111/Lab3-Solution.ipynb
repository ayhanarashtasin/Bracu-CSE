```python
#HW_Task-2
class Vector3D:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z
        print(f'Vector <{self.x}, {self.y}, {self.z}> created.')

    def magnitude(self):
        return (self.x**2 + self.y**2 + self.z**2)**0.5

    def dot_product(self, vctr):
        return self.x * vctr.x + self.y * vctr.y + self.z * vctr.z

    def cross_product(self, sclr):
        cross_x = self.y * sclr.z - self.z * sclr.y
        cross_y = self.z * sclr.x - self.x * sclr.z
        cross_z = self.x * sclr.y - self.y * sclr.x
        return Vector3D(cross_x, cross_y, cross_z)


V1 = Vector3D(2, -3, 1)
V2 = Vector3D(-1, 4, 0)
m_v1 = V1.magnitude()
m_v2 = V2.magnitude()
print(f'Magnitude of the first vector = {m_v1}')
print(f'Magnitude of the second vector = {m_v2}')
dot_p = V1.dot_product(V2)
print(f'Dot product of the two vectors = {dot_p}')
cross_p = V1.cross_product(V2)
print(f'Cross product of the two vectors = <{cross_p.x}, {cross_p.y}, {cross_p.z}>')
```

```python
 #HW-Task-6
class Store:
    def __init__(self, balance):
        print("New branch created!")
        self.balance = balance
        self.items = {}


    def add_item(self, items):
        name, stock, b_price, s_price = items
        if name not in self.items:
            self.items[name] = {'stock': stock,'buying_price': b_price,'selling_price': s_price }
            self.balance -= stock * b_price
            print(f"Item added: {name}")
        else:
            self.items[name]['stock'] += stock
            self.balance -= stock * b_price
            print(f"Restocked item: {name}, Current Stock: {self.items[name]['stock']}")

    def viewAllItems(self):
        list_items = ', '.join(self.items.keys())
        print(f"All Items: {list_items}")

    def viewAllItemDetails(self):
        print(self.items)

    def sell_item(self, name_1, number):
        if name_1 in self.items and self.items[name_1]['stock'] >= number:
            self.items[name_1]['stock'] -= number
            revenue = number * self.items[name_1]['selling_price']
            self.balance += revenue
            print(f"Sold {number} {name_1}, Revenue: {revenue}")
        else:
            print(f"Sorry! {name_1} is not available at your desired quantity. Currently we have: {self.items.get(name_1, {'stock': 0})['stock']}")

# Driver code
print("==========================")
branch1 = Store(5000)
print(f"Current Balance: {branch1.balance}")
print(f"Total items: {len(branch1.items)}")
branch1.viewAllItems()
branch1.viewAllItemDetails()
print("==========================")
print(f"Current Balance: {branch1.balance}")
branch1.add_item(["ChaCha Noodles", 10, 5, 8])
print(f"Current Balance: {branch1.balance}")
branch1.add_item(["Sparrow Shampoo", 5, 10, 20])
print(f"Current Balance: {branch1.balance}")
print("==========================")
branch1.viewAllItems()
print()
branch1.viewAllItemDetails()
print()
print("==========================")
print(f"Current Balance: {branch1.balance}\n")
branch1.sell_item("ChaCha Noodles", 15)
print(f"Current Balance: {branch1.balance}\n")
branch1.viewAllItemDetails()
print()
branch1.sell_item("ChaCha Noodles", 10)
print()
print(f"Current Balance: {branch1.balance}\n")
branch1.viewAllItemDetails()
print()
print("==========================")
print(f"Current Balance: {branch1.balance}\n")
branch1.add_item(["ChaCha Noodles", 5, 5, 8])
print()
branch1.viewAllItemDetails()
print()
print(f"Current Balance: {branch1.balance}\n")
print("==========================")
```
