```python
#Task-3
class Spaceship:
  def __init__(self,name,capacity):
    self.name = name
    self.capacity = capacity
    self.add = 0
    self.lst = []
  def load_cargo(self,obj):
    if obj.getweight()>self.capacity:
      print(f"Warning: Unable to load {obj.getname()} inside {self.name}.Exceeds capacity by {(obj.getweight()-self.capacity)+self.add}.")
    else:
      self.add+=obj.getweight()
    self.lst.append(obj.getname())
  def display_details(self):
    print(f"Spaceship Name: {self.name}")
    print(f"Capacity: {self.capacity}")
    print(f"Current cargo Weight: {self.add}")
    print(f"Cargo: {self.lst}")


class Cargo:
  def __init__(self,name,weight):
    self.__name = name
    self.__weight = weight
  def setname(self,name):
    if type(name)==str:
      self.__name = name
  def getname(self):
    return self.__name
  def setweight(self,weight):
    if weight>=0:
      self.__weight = weight
  def getweight(self):
    return self.__weight

# Creating spaceships
falcon = Spaceship("Falcon", 50000)
apollo = Spaceship("Apollo", 100000)
enterprise = Spaceship("Enterprise", 220000)
print("1.===================================")
# Creating cargo
gold = Cargo("Gold", 20000)
platinum = Cargo("Platinum", 25000)
dilithium = Cargo("Dilithium", 50000)
trilithium = Cargo("Trilithium", 70000)
neutronium = Cargo("Neutronium", 80000)
print("2.===================================")
# Loading cargo onto spaceships
falcon.load_cargo(gold)
falcon.load_cargo(platinum)
falcon.display_details()
print("3.===================================")
apollo.load_cargo(gold)  # Apollo will not reach its total capacity
apollo.display_details()
print("4.===================================")
falcon.load_cargo(neutronium)  # This should exceed Falcon's capacity
print("5.===================================")
enterprise.load_cargo(dilithium)
enterprise.load_cargo(trilithium)
enterprise.load_cargo(neutronium)  # This should not exceed Enterprise's capacity
enterprise.display_details()


```

```python
#Task-5
class Train:
  def __init__(self,name,*args):
    self.name = name
    self.start = args[0]
    self.destination = args[len(args)-1]
    self.lst = []
    self.a_dict = {}
    for i in args:
      self.lst.append(i)
    print(f"welcme abroad on {self.name} ")
    print(f"Start: {self.start}")
    print(f"Destination: {self.destination}")

  def addPassenger(self,*obj):
    for i in obj:
      print(f"{i.name} Welcome abroad")
      if i.x1 == "" and i.x2 =="":
        self.a_dict[i.name] = [self.start,self.destination]
      elif i.x2 =="":
        self.a_dict[i.name] = [i.x1,self.destination]
      else:
       self.a_dict[i.name] = [i.x1,i.x2]
  def allPassengerDetails(self):

    for key,value in self.a_dict.items():
      x = 0
      y = 0
      for j in range(len(self.lst)):
        if value[0]==self.lst[j]:
          x = j
        elif value[1]==self.lst[j]:
          y = j

      print(f"Name: {key},Start: {value[0]},Destination: {value[1]},Fair: ${(y-x)*100}")


class Passenger:
  def __init__(self,name,x1 = "",x2 = ""):
    self.name = name
    self.x1 = x1
    self.x2 = x2

t1 = Train('T1-Express','New York','Manhattan','Brooklyn','Boston')
print("1========================")
p1 =Passenger("Naruto")
t1.addPassenger(p1)
p2 = Passenger("Sasuke","Manhattan")
p3 = Passenger("Hinata","Manhattan","Brooklyn")
print("2========================")
t1.addPassenger(p2,p3)
print("3========================")
t1.allPassengerDetails()
print("4========================")
t2 = Train('Europe-Express','London','Paris','Brussels','Turkey')
print("5========================")
p4 =Passenger("Max","London","Brussels")
p5 = Passenger("Eleven","Paris")
p6 = Passenger("Mike","Brussels")
t2.addPassenger(p4,p5,p6)
print("6========================")
t2.allPassengerDetails()


```

```python
#Task-6
class BracuStudent:
  def __init__(self,name,home):
    self.name = name
    self.home = home
    self.buspass = False
  def get_pass(self):
    if self.buspass == False:
      self.buspass = True
  def show_details(self):
    print(f"Student Name: {self.name}")
    print(f"Lives in {self.home}")
    print(f"Have bus pass? {self.buspass}")
class BracuBus:
  def __init__(self,route,max = 2):
    self.route = route
    self.max = max
    self.passenger = []
  def board(self,*obj):
    if len(obj) != 0:
      for i in obj:
        if i.buspass == False:
          print("You don't have a bus pass!")
        elif i.home == self.route and i.buspass == True:
          self.passenger.append(i.name)
          print(f"{i.name} boarded the bus")
        else:
          print("You got on the wrong bus!")

        if len(self.passenger)==self.max:
          print("Bus is full!")
          break
    else:
        print("No passengers!")

  def show_details(self):
    print(f"Bus Route: {self.route}")
    print(f"Passengers Count: {len(self.passenger)} (Max: {self.max})")
    print(f"Passengers on Board: {self.passenger}")

st1 = BracuStudent("Afif", "Mirpur")
print("1===========================")
st2 = BracuStudent("Shanto", "Motijheel")
st3 = BracuStudent("Taskin", "Mirpur")
st1.show_details()
st2.show_details()
print("2===========================")
st3.show_details()
print("3===========================")
bus1 = BracuBus("Mirpur")
bus2 = BracuBus("Azimpur", 5)
bus1.show_details()
bus2.show_details()
print("4===========================")
st2.get_pass()
st3.get_pass()
print("5===========================")
st2.show_details()
st3.show_details()
print("6===========================")
bus1.board()
print("7===========================")
bus1.board(st1, st2)
print("8===========================")
st1.get_pass()
st2.home = "Mirpur"
st1.show_details()
st2.show_details()
print("9===========================")
bus1.board(st1, st2, st3)
print("10===========================")
bus1.show_details()

```
