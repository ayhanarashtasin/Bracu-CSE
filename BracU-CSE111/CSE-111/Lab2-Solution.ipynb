```python
#Task_1
def hospital(**kwargs):
  max_s = 0
  max_name = []
  for key,value in kwargs.items():
    if value>max_s:
      max_s=value
      max_name=[key]
    elif value == max_s:
      max_name.append(key)
  return max_s,max_name
max_s, max_name = hospital(Neymar = 2000, Dembele = 600, Reus = 500, Bale = 2000)
print(f"Highest fee was {max_s} tk")
print(f"which was paid by ")
for i in max_name:
  print(i,end=" ")
```

```python
#Task-3
def assign_students_to_sections(sections, *student_names):



  a_dict = {}
  for i in range(len(sections)):
    a_dict[sections[i]] = []


  for j in range(len(student_names)):
    sum = 0
    for k in student_names[j]:
      sum += ord(k)


    section_a = sum % len(sections)
    section_b = sections[section_a]


    a_dict[section_b].append(student_names[j])

  return a_dict
a_dict = assign_students_to_sections('ABCDE', 'Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace')

print(a_dict)
```

```python
def classroom(f_name, l_name, s_id, m_name=""):



    first = f_name[:3].upper()


    middle = m_name if m_name else ""


    last = l_name[-3:].lower()


    st_id = f"_{s_id % 10000}"


    value = f"{first}{middle}{last}{st_id}"

    return value


x=input("First Name: ")
y=input("Middle Name: ")
z=input("Last Name: ")
a=int(input("Student ID: "))
print(classroom(x, z, a))
print(classroom(x, z, a, y))
```

```python
def game(names):


  a_list = []

  for i  in range(len(names)):

    chr_1 = names[i][0].lower()
    chr_2 = names[i][-1].upper()


    chr3 = names[i][1:-1]
    chr4 = chr3[::-1]
    chr_ascii = ""
    for j in chr4:
      chr_ascii += str(ord(j))


    x  = f"{chr_1}{chr_ascii}{chr_2}"


    a_list.append(x)

  return a_list
names = ["Alex", "Bob", "Trudy"]
a_list = game(names)

print("Encrypted Keys:", a_list)
```

```python
import random
def playRockPaperScissor(rounds):

  score1 = 0
  score2 = 0

  x = ["rock", "paper", "scissor"]
  for i in range(rounds):

    action1 = input("Choose rock, paper, or scissor: ").lower()


    action2 = random.choice(x)


    print(f"Computer: {action2}")


    if action1 == action2:
      print("Tie round!")
    elif action1 == "rock" and action2 == "scissors":
      print("You win the round!")
      score1 += 1
    elif action1 == "paper" and action2 == "rock":
      print("You win the round!")
      score1 += 1
    elif action1 == "scissors" and action2 == "paper":
      print("You win the round!")
      score1 += 1
    else:
      print("Computer wins the round!")
      score2 += 1


    if score1 > score2:
       winner = "player"
    else:
      "computer"
    if score2 > score1:
       winner = "computer"
    else:
       "tie"

  return score1, score2, winner

rounds = int(input("Enter the number of rounds: "))

score1,score2, winner = playRockPaperScissor(rounds)

print(f"Your Score: {score1}")
print(f"Computer's Score: {score2}")

if winner == "player":
  print("You have won the game!")
elif winner == "computer":
  print("Computer has won the game!")
else:
  print("The game is a tie!")
```
